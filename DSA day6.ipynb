```python
num=int(input())
rev=' '
for i in str(num):
    rev=i+rev
rev=int(rev)
print(rev,type(rev))
```

```python
num=int(input())
temp=num
rev=0
while num>0:
    digit=num%10 
    rev=rev*10+digit
    num=num//10
print(rev)
if temp==rev:
    print(f'{temp} is palindrome number')
else:
    print(f'{temp} is not a palindrome number')
```

```python
num=int(input())
temp=str(num)
print(type(temp))
if temp==temp[::-1]:
    print("palindrome")
else:
    print("Not an palindrome")
```
