```python
# 1. Check Even or Odd Number
num=int(input())
if num%2==0:
    print(f'{num} is even')
else:
    print(f'{num} is odd')
```

```python
# 2. Print Even Numbers from 1 to N
n=int(input())
for i in range(1,n+1):
    print(i, end=' ')
```

```python
# 3. Print Odd Numbers from 1 to N
num=int(input())
for i in range(1,num+1):
    if i%2!=0:
        print(i, end=' ')
```

```python
# 4. Count Even Numbers in a List
numbers = [2, 5, 8, 11, 14, 17]
count=0
for num in numbers:
    if num%2==0:
        print(num)
        count=count+1
print("count:",count)

```
