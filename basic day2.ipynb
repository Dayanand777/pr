```python
# 2. Print Even Numbers from 1 to N
num=20
for i in range(1,num):
    if i%2==0:
        print(i)
```

```python
# 3. Print Odd Numbers from 1 to N
num=20
for i in range (1,num):
    if i%2!=0:
        print(i)
```

```python
# 4. Count Even Numbers in a List
numbers = [2, 5, 8, 11, 14, 17]
count=0
for x in numbers:
    if x%2==0:
        print(count)
        count+=1
print('count is',count)
```

```python
# 5. Sum of Odd Numbers from 1 to N
num=int(input("Enter a number:"))
sum=0
for i in range(1,num+1):
    if i%2!=0: # For ODD
    # if i%2==0: # For Even
        sum=sum+i
print(f'sum is: {sum}')
```

```python
# Problem 5 (Placement style)
# Check whether last digit is even or odd
num=int(input("Enter a number"))
digit=num%10
print("LAST DIGIT:",digit)
if digit%2==0:
    print(f'{digit} is Even')
else:
    print(f'{digit} is odd')
```
