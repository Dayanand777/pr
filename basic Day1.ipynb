```python
 # 1. Print Numbers from 1 to N
num=int(input("Enter a number:"))
for i in range(1,num+1):
    print(i,end=" ")
```

```python
# 2. SUM of N number.
num=int(input("Enter a number:"))
sum=0
for i in range(1,num+1):
    sum=sum+i
print(sum)
```

```python
# Swap two number
a=int(input("Enter first number:"))
b=int(input("Enter second number:"))
print(f"Before swap: a:{a}  b:{b}")
a,b=b,a
print(f"After swap: a:{a}  b:{b}")
```
