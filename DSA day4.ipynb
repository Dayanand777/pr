```python
# Prime Number
num=int(input("Enter a number"))
if num<=1:
    print("Not an prime")
else:
    for i in range(2,num):
        if num%i==0:
            print("Not an prime")
            break
    else:
        print("Prime number")        
```

```python
# Using recursion
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        return n*factorial(n-1)
# num=int(input())
n=int(input())
factorial(n)
# print(factorial(num))
```

```python
# Find sum of factorials from 1 to N
n=int(input())
temp=n
fact=1
sum=0
for i in range(1,n+1):
    fact=fact*i
    sum=sum+fact
    # print(fact)
print(sum)
```

```python
# Check if a number is strong number.
num=int(input())
temp=num
fact=1
sum=0
for i in str(num):
    fact=1
    for j in range(1,int(i)+1):
        fact=fact*j
    sum=sum+fact
    print(fact)
    
print(sum)
if sum==temp:
    print("Strong number")
else:    print("Not a strong number")  
```

```python
# Strong number
num=int(input())
fact=1
sum=0
for i in str(num):
    for j in range(1,int(i)+1):
        fact=fact*j
    sum=sum+fact
    print(fact)
    fact=1

print(sum)
```

```python
# * Triangle
num=int(input())
for i in range(num+1):
    for j in range(num-i):
        print(' ',end='')
    for k in range(2*i-1):
        print('*',end='')
    print()
        
        
```
