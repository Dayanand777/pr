```python
# Find Nth Fibonacci Number
# num=int(input())
# a=0
# b=1
# for i in range(num):
#     c=a+b
#     a=b
#     b=c
# print(a)


def fib(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a

n = int(input())
print(fib(n))
# • Time Complexity = O(n)
# • Space Complexity = O(1)
```

```python
'''
Fibonacci Using Recursion
Concept:
Function calls itself. 
'''
def fib(n):
    if n==0:
        return 0
    elif n==1:
        return 1
    else:
        return fib(n-1)+fib(n-2)
n=int(input())
print(fib(n))

# Note: This is slow (O(2^n)), used only for learning recursion.
```

```python
'''Problem 3
Check if a number is Fibonacci'''

def fib(n):
    a=0
    b=1
    while a<=n:
        if a==n:
            return True
        a=b
        b=a+b
    return False

n=int(input())
if fib(n):
    print("Fibonacci")
else:
    print("not")
    


# TRY
```
